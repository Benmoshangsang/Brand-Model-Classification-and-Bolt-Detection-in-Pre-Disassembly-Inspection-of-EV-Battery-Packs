```python
import json
from collections import defaultdict

# 设置路径：可改为 val.json 路径
json_path = '/root/autodl-tmp/coco_dataset/annotations/train.json'

# 定义要统计的合法类别（bolt1 到 bolt23）
target_classes = [f'bolt{i}' for i in range(1, 23)]  # bolt1~bolt23

with open(json_path, 'r') as f:
    data = json.load(f)

# 1. 构建类名与 ID 的映射
categories = data['categories']
name_to_id = {cat['name']: cat['id'] for cat in categories}
id_to_name = {cat['id']: cat['name'] for cat in categories}

# 2. 初始化计数器
count_dict = {name: 0 for name in target_classes}

# 3. 遍历标注，累加合法类别的数量
for ann in data['annotations']:
    cat_id = ann['category_id']
    class_name = id_to_name.get(cat_id, None)
    if class_name in count_dict:
        count_dict[class_name] += 1

# 4. 排序输出
print("📊 bolt1 到 bolt23 的标注框数量统计：\n")
for name in sorted(count_dict.keys(), key=lambda x: int(x[4:])):  # 按编号排序
    print(f"  {name:<7} -> {count_dict[name]} 个标注框")
```
