```python
import os
import json

def convert_jpg_to_JPG(image_dir, json_dir):
    count = 0
    for fname in os.listdir(image_dir):
        if fname.lower().endswith('.jpg'):
            base = os.path.splitext(fname)[0]
            old_path = os.path.join(image_dir, fname)
            new_path = os.path.join(image_dir, base + '.JPG')

            # 重命名图像文件（只变后缀大小写）
            os.rename(old_path, new_path)
            count += 1

            # 查找并更新对应的 JSON 文件
            json_path = os.path.join(json_dir, base + '.json')
            if os.path.exists(json_path):
                try:
                    with open(json_path, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                    data['imagePath'] = base + '.JPG'
                    with open(json_path, 'w', encoding='utf-8') as f:
                        json.dump(data, f, ensure_ascii=False, indent=2)
                except Exception as e:
                    print(f"❌ 更新失败: {json_path} — {e}")
            else:
                print(f"⚠️ JSON 文件缺失: {json_path}")
    print(f"✅ 完成：共更新 {count} 张图片和对应 JSON 文件")

if __name__ == '__main__':
    convert_jpg_to_JPG('/root/autodl-tmp/coco_dataset/images/train', '/root/autodl-tmp/coco_dataset/jsons/train')
    convert_jpg_to_JPG('/root/autodl-tmp/coco_dataset/images/val', '/root/autodl-tmp/coco_dataset/jsons/val')
```
