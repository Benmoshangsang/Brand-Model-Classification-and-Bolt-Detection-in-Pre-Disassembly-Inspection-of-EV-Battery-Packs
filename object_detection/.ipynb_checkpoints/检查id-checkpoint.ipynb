```python
import json

json_path = '/root/autodl-tmp/coco_dataset/annotations/train.json'
with open(json_path, 'r') as f:
    data = json.load(f)

print("✅ 顶层字段包括：", list(data.keys()))
print("✅ 每张图片的字段包括（示例第一张）：", list(data['images'][0].keys()))

errors = []
for img in data.get('images', []):
    if 'brand_id' not in img or not isinstance(img['brand_id'], int) or img['brand_id'] < 0:
        errors.append(img.get('file_name', 'Unknown'))

if errors:
    print("🚨 以下图片缺失 brand_id 或设置错误：")
    for fname in errors:
        print(" -", fname)
else:
    print("✅ 所有图片都设置了合法的 brand_id")
```
