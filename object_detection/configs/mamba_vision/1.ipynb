```python
import json
from collections import defaultdict

with open('/root/autodl-tmp/coco_dataset/annotations/train.json') as f:
    data = json.load(f)

ann_map = defaultdict(list)
for ann in data['annotations']:
    ann_map[ann['image_id']].append(ann)

empty_images = [img['id'] for img in data['images'] if len(ann_map[img['id']]) == 0]
print(f'⚠️ 有 {len(empty_images)} 张图片没有任何标注对象。')
```
