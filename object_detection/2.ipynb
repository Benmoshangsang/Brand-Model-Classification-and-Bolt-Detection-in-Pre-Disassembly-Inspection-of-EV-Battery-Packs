```python
import json
import os

# 两个 COCO 标注文件路径
json_paths = {
    'train': '/root/autodl-tmp/coco_dataset/annotations/train.json',
    'val': '/root/autodl-tmp/coco_dataset/annotations/val.json'
}

for split, json_path in json_paths.items():
    print(f'\n📁 正在检查 {split}.json 是否包含 bolt16 标注框...')
    
    if not os.path.exists(json_path):
        print(f'❌ 未找到文件: {json_path}')
        continue

    with open(json_path, 'r') as f:
        data = json.load(f)

    # 类别映射
    categories = data.get('categories', [])
    category_name_to_id = {cat['name']: cat['id'] for cat in categories}
    image_id_to_name = {img['id']: img['file_name'] for img in data['images']}

    # 检查是否存在 bolt27
    if 'bolt16' not in category_name_to_id:
        print(f"✅ {split}.json 中不包含类别 'bolt16'")
        continue

    bolt16_id = category_name_to_id['bolt16']

    # 查找包含 bolt27 的图像
    bolt16_image_ids = set(
        ann['image_id'] for ann in data['annotations'] if ann['category_id'] == bolt16_id
    )
    bolt16_file_names = [image_id_to_name[iid] for iid in bolt16_image_ids if iid in image_id_to_name]

    if bolt16_file_names:
        print(f"⚠️ 发现 {len(bolt16_file_names)} 张图包含 bolt16：")
        for name in bolt16_file_names:
            print(f"  📸 {name}")
    else:
        print(f"✅ {split}.json 中未发现含有 bolt16 的图像")
```
