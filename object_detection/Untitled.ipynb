```python
import json
import os

# 输入两个 JSON 文件路径
json_paths = {
    'train': '/root/autodl-tmp/coco_dataset/annotations/train.json',
    'val': '/root/autodl-tmp/coco_dataset/annotations/val.json'
}

for split, path in json_paths.items():
    print(f"\n📂 正在检查 {split}.json 中包含 bolt23 的图片...")

    if not os.path.exists(path):
        print(f"❌ 未找到文件: {path}")
        continue

    with open(path, 'r') as f:
        data = json.load(f)

    # 构建类别映射
    categories = data['categories']
    name_to_id = {cat['name']: cat['id'] for cat in categories}
    id_to_name = {cat['id']: cat['name'] for cat in categories}
    image_id_to_name = {img['id']: img['file_name'] for img in data['images']}

    # 确认 bolt23 是否存在
    if 'bolt23' not in name_to_id:
        print("⚠️ 类别 'bolt23' 不存在，跳过。")
        continue

    bolt23_id = name_to_id['bolt23']

    # 获取所有包含 bolt23 的 image_id
    bolt23_image_ids = set(
        ann['image_id'] for ann in data['annotations'] if ann['category_id'] == bolt23_id
    )

    # 映射为文件名
    bolt23_filenames = [
        image_id_to_name[img_id] for img_id in bolt23_image_ids if img_id in image_id_to_name
    ]

    # 输出结果
    if bolt23_filenames:
        print(f"✅ 发现 {len(bolt23_filenames)} 张图片包含 bolt23 标注框：\n")
        for name in sorted(bolt23_filenames):
            print(f"  📸 {name}")
    else:
        print("✅ 未发现包含 bolt23 的图片。")
```
