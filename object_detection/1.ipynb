```python
import json

# 替换为你自己的标注文件路径
json_path = '/root/autodl-tmp/coco_dataset/annotations/train.json'

with open(json_path, 'r') as f:
    data = json.load(f)

# 提取类别映射
categories = data.get('categories', [])
category_name_to_id = {cat['name']: cat['id'] for cat in categories}
id_to_name = {cat['id']: cat['name'] for cat in categories}

# 输出所有类别及其 ID
print("📌 所有类别列表:")
for name, id_ in category_name_to_id.items():
    print(f'  ID={id_:>2} | class="{name}"')

# ✅ 检查是否存在 bolt27 和 bolt17
for target in ['bolt27', 'bolt17']:
    if target in category_name_to_id:
        print(f'\n✅ 存在类别 "{target}"，对应 ID: {category_name_to_id[target]}')
    else:
        print(f'\n❌ 未找到类别 "{target}"')

# ✅ 额外：检查标注中是否引用了这些类别
ann_category_ids = set(ann['category_id'] for ann in data['annotations'])
ann_category_names = [id_to_name[cid] for cid in ann_category_ids if cid in id_to_name]

print(f"\n📊 当前数据集中实际用到的类别有 {len(ann_category_names)} 类：")
print(sorted(ann_category_names))
```
