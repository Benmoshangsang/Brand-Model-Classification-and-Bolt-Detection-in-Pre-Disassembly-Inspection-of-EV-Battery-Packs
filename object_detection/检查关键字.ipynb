```python
import json

def inspect_coco_json(json_path):
    print(f"\n🔍 正在检查文件: {json_path}")
    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # 检查 images 字段
    print("\n📁 [images] 中的字段:")
    if 'images' in data and isinstance(data['images'], list):
        image_keys = set()
        for img in data['images']:
            image_keys.update(img.keys())
        print("字段集合:", image_keys)
        brand_image_missing = sum(1 for img in data['images'] if 'brand_id' not in img)
        print(f"⚠️ 缺失 brand_id 的 images 条目数: {brand_image_missing}")
    else:
        print("❌ 未找到 images 字段")

    # 检查 annotations 字段
    print("\n📁 [annotations] 中的字段:")
    if 'annotations' in data and isinstance(data['annotations'], list):
        anno_keys = set()
        for anno in data['annotations']:
            anno_keys.update(anno.keys())
        print("字段集合:", anno_keys)
        brand_anno_missing = sum(1 for anno in data['annotations'] if 'brand_id' not in anno)
        print(f"⚠️ 缺失 brand_id 的 annotations 条目数: {brand_anno_missing}")
    else:
        print("❌ 未找到 annotations 字段")

    # 类别字段
    print("\n📁 [categories] 类别数:", len(data.get('categories', [])))

# 路径配置
paths = {
    'train': '/root/autodl-tmp/coco_dataset/annotations/train.json',
    'val': '/root/autodl-tmp/coco_dataset/annotations/val.json'
}

# 执行检查
for name, path in paths.items():
    inspect_coco_json(path)
```
