```python
import os
import json

# 设置两个待处理目录
json_dirs = [
    '/root/autodl-tmp/coco_dataset/jsons/train',
    '/root/autodl-tmp/coco_dataset/jsons/val'
]

# 目标替换：bolt27 → bolt19
old_label = 'bolt27'
new_label = 'bolt19'

# 总体统计
total_file_modified = 0
total_object_modified = 0

for json_dir in json_dirs:
    print(f'\n📂 正在处理目录：{json_dir}')
    
    dir_file_modified = 0
    dir_object_modified = 0

    for filename in os.listdir(json_dir):
        if not filename.endswith('.json'):
            continue

        file_path = os.path.join(json_dir, filename)

        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
        except Exception as e:
            print(f'❌ 读取失败: {file_path}，错误：{e}')
            continue

        modified = False
        count_in_file = 0

        for shape in data.get('shapes', []):
            if shape.get('label') == old_label:
                shape['label'] = new_label
                modified = True
                count_in_file += 1

        if modified:
            with open(file_path, 'w', encoding='utf-8') as f:
                json.dump(data, f, ensure_ascii=False, indent=2)

            dir_file_modified += 1
            dir_object_modified += count_in_file
            print(f'✅ 修改文件: {filename}（替换 {count_in_file} 个标注）')

    # 每个目录处理后汇总
    total_file_modified += dir_file_modified
    total_object_modified += dir_object_modified
    print(f'\n📊 目录 {os.path.basename(json_dir)} 处理结果：')
    print(f'   📄 修改文件数：{dir_file_modified}')
    print(f'   🔧 替换标注框数：{dir_object_modified}')

# 总体输出
print('\n✅ 全部处理完成！')
print(f'📄 总计修改文件数：{total_file_modified}')
print(f'🔧 总计替换标注框数：{total_object_modified}')
```
