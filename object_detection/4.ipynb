```python
import json
import os

# 两个标注文件路径（可根据实际路径修改）
json_paths = {
    'train': '/root/autodl-tmp/coco_dataset/annotations/train.json',
    'val': '/root/autodl-tmp/coco_dataset/annotations/val.json'
}

for split, path in json_paths.items():
    print(f"\n📂 正在检查 {split}.json 中包含 bolt23 的图片...")

    if not os.path.exists(path):
        print(f"❌ 文件未找到: {path}")
        continue

    with open(path, 'r') as f:
        data = json.load(f)

    # 构建类名 ↔ id 映射
    categories = data['categories']
    name_to_id = {cat['name']: cat['id'] for cat in categories}
    id_to_name = {cat['id']: cat['name'] for cat in categories}

    # 图像信息：id → file_name 和 brand_id
    image_info_map = {
        img['id']: {'file_name': img['file_name'], 'brand_id': img.get('brand_id', -1)}
        for img in data['images']
    }

    # 检查是否有 bolt23 类别
    if 'bolt23' not in name_to_id:
        print("⚠️ 类别 'bolt23' 不存在，跳过。")
        continue

    bolt23_id = name_to_id['bolt23']

    # 统计使用 bolt23 的 image_id
    bolt23_image_ids = set(
        ann['image_id'] for ann in data['annotations'] if ann['category_id'] == bolt23_id
    )

    # 输出结果
    if not bolt23_image_ids:
        print("✅ 未发现包含 bolt23 的图片。")
        continue

    print(f"✅ 发现 {len(bolt23_image_ids)} 张图片包含 bolt23 标注框：\n")
    for img_id in sorted(bolt23_image_ids):
        if img_id in image_info_map:
            info = image_info_map[img_id]
            print(f"  📸 {info['file_name']}   | brand_id={info['brand_id']}   | split={split}")
```
